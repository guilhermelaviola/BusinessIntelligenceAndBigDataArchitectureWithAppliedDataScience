# **Data Mining Techniques & Algorithms**
Data Mining is vital for organizations to extract meaningful insights from large datasets by uncovering patterns, relationships, and anomalies that support better decision-making. It includes a variety of techniques such as classification algorithms (e.g., Decision Trees, SVM, KNN) for predicting and categorizing data, association rule mining (e.g., Apriori and FP-Growth) for identifying relationships between items, and anomaly detection methods (e.g., Isolation Forest, One-Class SVM, DBScan) for spotting unusual or risky behavior. These methods are widely applied in areas like marketing, recommendation systems, fraud detection, cybersecurity, healthcare, inventory optimization, and customer segmentation. Overall, Data Mining transforms raw data into actionable knowledge, helping businesses optimize operations, personalize strategies, reduce risks, and gain a competitive edge.

```python
# Importing all the necessary libraries:
import pandas as pd
from mlxtend.frequent_patterns import fpgrowth, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

## **Example: FP-Growth algorithm**
The example below shows and example of data analysis workflow, from data loading to analysis of the results obtained with the FP-Growth algorithm. By analyzing the results, we can identify purchasing patterns, such as the tendency to buy milk and bread together.

```python
# Importing and displaying the DataFrame:
url = 'https://raw.githubusercontent.com/Sarvandani/Machine_learning_Association-rule-learning_Market-Basket-Analysis/refs/heads/main/Groceries_dataset.csv'
df = pd.read_csv(url)
df.head()
```

```python
# Filtering the rows with Member_number = 1808:
df[df['Member_number'] == 1808]
```

```python
# Creating a new Dataframe to group the filtered result above by 'Date':
grouping = df[df['Member_number'] == 1808].groupby('Date')['itemDescription'].apply(list)
grouping.head()
```

```python
# Creating a new Dataframe to group the filtered result above by 'Date':
grouped_transactions = df.groupby(['Member_number', 'Date'])['itemDescription'].apply(list).reset_index(name='items')['items']
grouped_transactions.head()
```

```python
# Instantiating the transaction encoder:
te = TransactionEncoder()

# Transforming the data into the adequate format:
te_ary = te.fit(grouped_transactions).transform(grouped_transactions)
df_te = pd.DataFrame(te_ary, columns=te.columns_)
df_te.head()
```

```python
# Applying the FP-Growth algorithm:
frequent_itemsets = fpgrowth(df_te, min_support=0.005, use_colnames=True)

# Displaying the most frequent item collections:
print(frequent_itemsets)
```

```python
# Generating a DataFrame of association rules with the 'score' metric:
rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.005)
soda_rules = rules[rules['antecedents'].apply(lambda x: 'soda' in x)]
soda_rules
```

```python
# Generating a sorted DataFrame of association rules with the 'score' metric:
soda_rules_sorted = soda_rules.sort_values(by='confidence', ascending=False)
soda_rules_sorted
```
