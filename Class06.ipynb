# **Introduction to Applied Statistics**
Applied Statistics is essential for analyzing and interpreting data, transforming raw information into actionable insights, and supporting informed decision-making across fields like business, healthcare, engineering, and social sciences. It involves key concepts such as measures of central tendency (mean, median, mode) and dispersion (range, standard deviation), which summarize data and reveal variability. Probability and probability distributions help assess uncertainty and model random events, while sampling techniques allow inferences about larger populations. Hypothesis testing evaluates evidence for or against assumptions, and parameter estimation uses sample data to estimate population characteristics. Overall, Applied Statistics equips us to identify patterns, test hypotheses, and make evidence-based decisions in a wide range of practical applications.

## **Practical Activity**
A cosmetics company wants to test the effectiveness of a new anti-wrinkle cream on a group of volunteers. The volunteers were randomly divided into two groups: one group that received the anti-wrinkle cream and another group that received a placebo. After 30 days of use, the depth of each volunteer's wrinkles was measured. Based on the wrinkle depth data, determine if the anti-wrinkle cream is truly effective in reducing wrinkles. Which hypothesis test would be most appropriate to analyze the data in this scenario and how would you apply it?

- Student's t-test for independent samples.
- Student's t-test for paired samples.
- Chi-square test of independence.
- One-way ANOVA test.
- Two-way ANOVA test.

The most appropriate test would be a) Student's t-test for independent samples, which compares the means of two independent groups to determine if there is a statistically significant difference between them. The outcome variable would be wrinkle depth, which is quantitative, i.e., measured numerically.

The test would be applied as follows:
### **Hypothesis definition**
- Null hypothesis: There is no significant difference in the average wrinkle depth between the group that used the anti-wrinkle cream and the group that used the placebo.
- Alternative hypothesis: The group that used the anti-wrinkle cream has a significantly lower average wrinkle depth than the group that used the placebo.
### **Data collection and preparation**
In real life, the data would need to be collected from women willing to participate in the test in person. There would be 3 columns of data: the volunteer's number, their randomly assigned group (Cream or Placebo), and the wrinkle depth after 30 days of testing. Since this is an activity, I generated the data randomly for all columns using Python in step 4.
### **Verification of assumptions**
Before running the t-test, the following must be ensured:
- Normality: Checking if the wrinkle depth follows a normal distribution within each group using the Shapiro-Wilk test or the histogram or Q-Q plot.
- Equality of variance: Using Levene's test or the F-test to check if the two groups have similar variances. If the variances are unequal, Welch's t-test, which is a variation of the independent t-test, should be used.
### **Test execution**
The Python code follows:

```python
# Setting random seed for reproducibility:
np.random.seed(42)

# Setting the number of violunteers:
num_volunteers = 100

# Generating the volunteers IDs:
volunteers = np.arange(1, num_volunteers + 1)

# Randomly assigning groups (50% Cream, 50% Placebo):
groups = np.random.choice(['Cream', 'Placebo'], size=num_volunteers)

# Generating wrinkle depth values (between 0 and 1):
wrinkle_depths = np.round(np.random.uniform(0, 1, num_volunteers), 3)

# Generating the DataFrame:
df = pd.DataFrame({
    'Volunteers': volunteers,
    'Group': groups,
    'Wrinkle Depth': wrinkle_depths
})

# Displaying first few rows:
print(df.head())
```

### **Interpretation of results**
- If p < 0.05: The difference in mean wrinkle depth is statistically significant. The cream is likely effective in reducing wrinkles.
- If p  0.05: There is no significant evidence that the cream reduces wrinkles compared to placebo.

In this case, no significant difference was found.

```python
# Separating data into two groups:
cream_group = df[df['Group'] == 'Cream']['Wrinkle Depth']
placebo_group = df[df['Group'] == 'Placebo']['Wrinkle Depth']

# Checking normality (Shapiro-Wilk test):
shapiro_cream = stats.shapiro(cream_group)
shapiro_placebo = stats.shapiro(placebo_group)
print(f'Shapiro-Wilk Test (Cream group): p-value = {shapiro_cream.pvalue:.4f}')
print(f'Shapiro-Wilk Test (Placebo group): p-value = {shapiro_placebo.pvalue:.4f}')

# Checking variance equality (Levene's test):
levene_test = stats.levene(cream_group, placebo_group)
print(f'Levene Test for Equal Variance: p-value = {levene_test.pvalue:.4f}')

# Performing the independent t-test:
equal_var = levene_test.pvalue > 0.05
t_stat, p_value = stats.ttest_ind(cream_group, placebo_group, equal_var=equal_var)

# Displaying the results:
print(f'T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}')

# Interpreting the results:
alpha = 0.05
if p_value < alpha:
    print('The anti-wrinkle cream significantly reduces wrinkles.')
else:
    print('No significant differences found.')
```
