# **Introduction to Applied Statistics**
Applied Statistics is essential for analyzing and interpreting data, transforming raw information into actionable insights, and supporting informed decision-making across fields like business, healthcare, engineering, and social sciences. It involves key concepts such as measures of central tendency (mean, median, mode) and dispersion (range, standard deviation), which summarize data and reveal variability. Probability and probability distributions help assess uncertainty and model random events, while sampling techniques allow inferences about larger populations. Hypothesis testing evaluates evidence for or against assumptions, and parameter estimation uses sample data to estimate population characteristics. Overall, Applied Statistics equips us to identify patterns, test hypotheses, and make evidence-based decisions in a wide range of practical applications.

## **Example: Normal Distribution**
The following example show how to apply statistics to a practical example in Python, by showing Normal Dsitribution and Linear Regression.

```python
# Downloading the DataFrame:
!gdown 1uZUL9YG02LIxRx0zG8UCcfWjrVpn2d7e
```

```python
# Importing and displaying the DataFrame:
df = pd.read_csv('SOCR-HeightWeight.csv')
df.head()
```

```python
# Doing new calculations with the data:
df['Height(Meters)'] = df['Height(Inches)'] * 0.0254
df['Weight(Kilos)'] = df['Weight(Pounds)'] * 0.4535
df.head()
```

```python
# Displaying the formatted DataFrame:
df = df.rename(columns = {'Height (Inches)': 'Height (meters)', 'Weight (Pounds)': 'Weight (kilos)'})
df.head()
```

```python
# Displaying detailed statistical data of the DataFrame:
df.describe()
```

```python
# Displaying Normal Distribution for Height (Meters):
sns.set(style='whitegrid')
plt.figure(figsize=(10, 6))
sns.histplot(df['Height(Meters)'], kde=True, color='skyblue', bins=30, stat='density')
plt.title('Normal Distribution - Height (Meters)')
plt.xlabel('Values')
plt.ylabel('Density')
plt.show()
```

```python
# Displaying Normal Distribution for Weight (Kilos):
sns.set(style='whitegrid')
plt.figure(figsize=(10, 6))
sns.histplot(df['Weight(Kilos)'], kde=True, color='skyblue', bins=30, stat='density')
plt.title('Normal Distribution - Weight (Kilos)')
plt.xlabel('Values')
plt.ylabel('Density')
plt.show()
```

```python
# Displaying Scatter Plot (Height x Weight) with Linear Regression:
plt.figure(figsize=(10, 6))
sns.regplot(x='Height(Meters)', y='Weight(Kilos)', data=df, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('Scatter Plot (Height x Weight) with Linear Regression')
plt.xlabel('Height (Meters)')
plt.ylabel('Weight (Kilos)')
plt.show()
```
